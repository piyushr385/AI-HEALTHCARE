```python
#import the necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings("ignore")
import joblib
```

```python
#load the dataset

df = pd.read_csv("cancer.csv")
df.drop(df.columns[[0,-1]], axis=1, inplace=True)
df.head(5)
```

```python
#describe the data

df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
# check the count of categorical feature as malignant,benign

df["diagnosis"].value_counts()
```

```python
sns.countplot(df.diagnosis)
plt.show()
```

```python
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

```python
df.columns
```

```python
le = LabelEncoder()
df["diagnosis"] = le.fit_transform(df["diagnosis"])
df["diagnosis"].unique()
df.head(5)
```

```python
#split the features
cols=['diagnosis','perimeter_mean','radius_mean','compactness_mean','concave points_mean',
                    'radius_se','perimeter_se','radius_worst','perimeter_worst','compactness_worst',
                    'concave points_worst','compactness_se','concave points_se','texture_worst','area_worst']
X = df.drop(cols, axis = 1 ) 
y = df['diagnosis']
```

```python
X.columns
```

```python
X.head(5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=43)
```

```python
print('Shape training set: X:{}, y:{}'.format(X_train.shape, y_train.shape))
print('Shape test set: X:{}, y:{}'.format(X_test.shape, y_test.shape))
```

```python
from sklearn.ensemble import RandomForestClassifier

cancer_model= RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 42)
cancer_model.fit(X_train, y_train)
cancer_model.score(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = cancer_model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.title('Heatmap of Confusion Matrix', fontsize = 15)
sns.heatmap(cm, annot = True)
plt.show()
```

```python
import pickle
# open a file, where you ant to store the data
file = open('cancer_model.pkl', 'wb')

# dump information to that file
pickle.dump(cancer_model, file)
```

```python
loaded_model = joblib.load('cancer_model.pkl')
result = loaded_model.score(X_test, y_test)
print(result)
```
